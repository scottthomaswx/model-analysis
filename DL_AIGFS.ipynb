```python
%%capture
!pip install openpyxl
!pip install pandas
!pip install requests
!pip install dateutil
!pip install numpy
!pip install os
!pip install datetime

from openpyxl import load_workbook
from openpyxl import Workbook
from openpyxl import utils as utils
from itertools import islice
import pandas as pd
from dateutil.parser import *
from datetime import datetime, timedelta
import numpy
import os
import requests

!rm -rf ./sample_data
!mkdir ./DATA
```

```python
wb = load_workbook('scott_Dec25.xlsx')
```

```python
direc = 'Dec25_AIGFS'
```

```python
ws = wb.active
```

```python
data = ws.values
```

```python
cols = next(data)[1:]
data = list(data)
idx = [r[0] for r in data]
data = (islice(r, 1, None) for r in data)
df = pd.DataFrame(data, index=idx, columns=cols)
```

```python
sixhr = ['00','06','12','18']
ilist = [1,3,5,7,9]
for i, row in df.iterrows():
    dirname=row.values[0][:6] + '_' + row.values[0][7:9]
    path=('./DATA/' + direc + '/' + dirname)
    if not os.path.exists(path):
        os.makedirs(path)
    for j in row[:]:
        date='20' + j[0:6]
        hour=j[7:9]
        fhour=j[-3:]
        if fhour == '000' and hour in sixhr:


            url1 = f'https://nomads.ncep.noaa.gov/pub/data/nccf/com/aigfs/prod/aigfs.{date}/{hour}/model/atmos/grib2/aigfs.t{hour}z.pres.f{fhour}.grib2'
            r1 = requests.get(url1)
            file1='./DATA/' + direc + '/' + dirname + '/'+ date + '_aigfs.t' + hour + 'z.pres.f' + fhour + '.grb2'
            with open(file1, 'wb') as f1:
                f1.write(r1.content)

            url2 = f'https://nomads.ncep.noaa.gov/pub/data/nccf/com/aigfs/prod/aigfs.{date}/{hour}/model/atmos/grib2/aigfs.t{hour}z.sfc.f{fhour}.grib2'
            r2 = requests.get(url2)
            file2='./DATA/' + direc + '/' + dirname + '/'+ date + '_aigfs.t' + hour + 'z.sfc.f' + fhour + '.grb2'
            with open(file2, 'wb') as f2:
                f2.write(r2.content)

        if hour in sixhr and i in ilist:

            url1 = f'https://nomads.ncep.noaa.gov/pub/data/nccf/com/aigfs/prod/aigfs.{date}/{hour}/model/atmos/grib2/aigfs.t{hour}z.pres.f{fhour}.grib2'
            r1 = requests.get(url1)
            file1='./DATA/' + direc + '/' + dirname + '/'+ date + '_aigfs.t' + hour + 'z.pres.f' + fhour + '.grb2'
            with open(file1, 'wb') as f1:
                f1.write(r1.content)

            url2 = f'https://nomads.ncep.noaa.gov/pub/data/nccf/com/aigfs/prod/aigfs.{date}/{hour}/model/atmos/grib2/aigfs.t{hour}z.sfc.f{fhour}.grib2'
            r2 = requests.get(url2)
            file2='./DATA/' + direc + '/' + dirname + '/'+ date + '_aigfs.t' + hour + 'z.sfc.f' + fhour + '.grb2'
            with open(file2, 'wb') as f2:
                f2.write(r2.content)
```
