```python
wb = load_workbook('scott_Dec25.xlsx')
```

```python
direc = 'Dec25_GFS'
```

```python
ws = wb.active
```

```python
data = ws.values
```

```python
cols = next(data)[1:]
data = list(data)
idx = [r[0] for r in data]
data = (islice(r, 1, None) for r in data)
df = pd.DataFrame(data, index=idx, columns=cols)
```

```python
sixhr = ['00','06','12','18']
ilist = [1,3,5,7,9]
for i, row in df.iterrows():
    dirname=row.values[0][:6] + '_' + row.values[0][7:9]
    path=('./DATA/' + direc + '/' + dirname)
    if not os.path.exists(path):
        os.makedirs(path)
    for j in row[:]:
        date='20' + j[0:6]
        hour=j[7:9]
        fhour=j[-3:]
        if fhour == '000' and hour in sixhr:

            url1 = 'https://noaa-gfs-bdp-pds.s3.amazonaws.com/gfs.' + date + '/' + hour +\
            '/atmos/gfs.t' + hour + 'z.pgrb2.0p25.f' + fhour
            r1 = requests.get(url1)
            file1='./DATA/' + direc + '/' + dirname + '/'+ date + '_gfs.t' + hour + 'z.pgrb2s.0p25.f' + fhour + '.grb2'
            with open(file1, 'wb') as f1:
                f1.write(r1.content)

        if hour in sixhr and i in ilist:

            url1 = 'https://noaa-gfs-bdp-pds.s3.amazonaws.com/gfs.' + date + '/' + hour +\
            '/atmos/gfs.t' + hour + 'z.pgrb2.0p25.f' + fhour
            r1 = requests.get(url1)
            file1='./DATA/' + direc + '/' + dirname + '/'+ date + '_gfs.t' + hour + 'z.pgrb2.0p25.f' + fhour + '.grb2'
            with open(file1, 'wb') as f1:
                 f1.write(r1.content)
```
